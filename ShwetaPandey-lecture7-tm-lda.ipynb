# Data Cleaning

```python
# Drop the columns not needed

reviews = reviews.drop(columns=['id', 'asins', 'dimension', 'ean', 'flavors', 'manufacturerNumber', 
                                'reviews.id', 'reviews.numHelpful', 'reviews.sourceURLs', 'reviews.userCity', 
                                'reviews.userProvince', 'sizes', 'sourceURLs', 'upc', 'weight'], axis=1).sample(2890)

# print head

reviews.head()
```

# Remove punctuation/lower casing

```python
# import regular expression library

import re 

# remove punctuations

reviews['reviews.text_processed'] = \
reviews['reviews.text'].map(lambda x: re.sub('[,\.!?]', '', str(x)))

# convert titles into lower case

reviews['reviews.text_processed'] = \
reviews['reviews.text_processed'].map(lambda x: x.lower())

# print head

reviews['reviews.text_processed'].head(20)
```

# Data preparation for LDA Analysis

```python
# install gensim
# pip install gensim
```

```python
# import libraries

import gensim
from gensim.utils import simple_preprocess
import nltk
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])

def sent_to_words(sentences):
    for sentence in sentences:
        # deacc=True removes punctuations
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))
        
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) 
             if word not in stop_words] for doc in texts]

data = reviews['reviews.text_processed'].values.tolist()
data_words = list(sent_to_words(data))
# remove stop words
data_words = remove_stopwords(data_words)
print(data_words[:1][0][:30])

```

```python
import gensim.corpora as corpora

# Create Dictionary

id2word = corpora.Dictionary(data_words)

# Create Corpus

texts = data_words

# Term Document Frequency

corpus = [id2word.doc2bow(text) for text in texts]

# View

print(corpus[:1][0][:30])
```

# LDA Model Training

```python
from pprint import pprint

# choosing number of topics

num_topics = 50

# Build LDA model

lda_model = gensim.models.LdaMulticore( corpus = corpus,
                                       id2word = id2word,
                                       num_topics = num_topics)

# print keywords in the topics

pprint(lda_model.print_topics())
doc_lda = lda_model[corpus]
```
